```python
import numpy as np
import seaborn as sns
import pylab as plt
from scipy.ndimage.measurements import labeled_comprehension
from scipy.ndimage.filters import gaussian_filter
from scipy import stats
from sklearn import preprocessing
%matplotlib inline
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('pdf', 'svg')
```

```python
plt.figure(figsize=(18,2))
signal = np.random.rand(400).reshape(-1, 1)
plt.plot(signal)
parcells = np.zeros(signal.shape)
n_parcells = 20
for i in range(n_parcells):
    parcells[i*len(signal)/n_parcells:(i+1)*len(signal)/n_parcells] = float(i)/float(n_parcells)
    
plt.plot(parcells)

between_parcel_variance = labeled_comprehension(signal,
                                     parcells,
                                     np.unique(parcells),
                                     np.mean,
                                     float,
                                     np.nan).var()
within_parcel_variance = labeled_comprehension(signal,
                                               parcells,
                                               np.unique(parcells),
                                               np.var,
                                               float,
                                               np.nan).mean()

print "overall signal variance = %g"%signal.var()
print "between parcell variance = %g"%between_parcel_variance
print "within parcell variance = %g"%within_parcel_variance
print "between/within variance ratio = %g"%(between_parcel_variance/within_parcel_variance)
```

```python
plt.figure(figsize=(18,2))
smooth_signal = preprocessing.MinMaxScaler().fit_transform(gaussian_filter(signal, 200))
plt.plot(smooth_signal)
plt.plot(parcells)

between_parcel_variance = labeled_comprehension(smooth_signal,
                                     parcells,
                                     np.unique(parcells),
                                     np.mean,
                                     float,
                                     np.nan).var()
within_parcel_variance = labeled_comprehension(smooth_signal,
                                               parcells,
                                               np.unique(parcells),
                                               np.var,
                                               float,
                                               np.nan).mean()

print "overall signal variance = %g"%smooth_signal.var()
print "between parcell variance = %g"%between_parcel_variance
print "within parcell variance = %g"%within_parcel_variance
print "between/within variance ratio = %g"%(between_parcel_variance/within_parcel_variance)
```
